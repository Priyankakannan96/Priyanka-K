# **`Break Statement`**


```python
# example that uses break statement in a for loop

# take range(10) and print 'The number is' + value
# break when num equals 5
# Loop through range(10)
for num in range(10):
    print(f"The number is {num}")
    if num == 5:
        break  # Exit the loop when num equals 5
```

# **`Continue Statement`**


```python
# Using same `for loop program` as in Break Statement section above
# Use a continue statement rather than a break statement

# take range(10) and print 'The number is' + value
# continue when num equals 5
```

```python
# Loop through range(10)
for num in range(10):
    if num == 5:
        continue  # Skip the rest of the loop when num equals 5
    print(f"The number is {num}")
```

Find a substring in string using `find` and `index` function

```python
# .find() => if present it will return starting index, not found then it will return -1
# .index() => if present it will return starting index, not found then it will give error
# Using .find()
find_position = sentence.find('data')  # This will return 38, as 'data' starts at index 38
not_found_find = sentence.find('python')  # This will return -1, as 'python' is not present

# Using .index()
try:
    index_position = sentence.index('data')  # This will return 38
    not_found_index = sentence.index('python')  # This will raise an error
except ValueError:
    not_found_index = "Substring not found"

# Print the results
print(f"Using .find(), position of 'data': {find_position}")  # 38
print(f"Using .find(), position of 'python': {not_found_find}")  # -1

print(f"Using .index(), position of 'data': {index_position}")  # 38
print(f"Using .index(), position of 'python': {not_found_index}")  # Substring not foundion of 'data' is: {index_position}")
```

# ***`Positional Arguments`***
Most arguments are identified by their position in function call
* Say `print(x,y)` will give different results from `print(y,x)`

What ever sequence is given while defining a function values must be taken in that sequence only
* Otherwise use argument name **`(keyword arguments)`** to take values
* We first define `positional argument` and then `keyword arguments`


```python
## Create substraction_function(small_number,large_number) and return difference between large_number and small_number

def subtraction_function(small_number, large_number):
    # Calculating the difference
    difference = large_number - small_number

    # Returning the result
    return difference
```

```python
# pass arguments in right order
def subtraction_function(small_number, large_number):
    # Calculating the difference
    difference = large_number - small_number

    # Returning the result
    return difference

# Calling the function with arguments in the correct order
result = subtraction_function(3, 10)  # Here 3 is the small number and 10 is the large number

# Printing the result
print(result)
```

```python
# always pass arguments using there name(keyword arguments) then order does not matter


result = subtraction_function(small_number=3, large_number=10)

# Printing the result
print(result)
```
